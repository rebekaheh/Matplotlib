### Part 1 - Macho Merging

* You will likely need to perform three different merges over the course of this activity, changing the names of your columns as you go along.

```python
# Import the necessary modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Bring each CSV into a separate data frame and read 'em
wwe2013_df = pd.read_csv(r"C:\Users\RebekahDSK\Desktop\Matplotlib\Winner Wrestling Project\CSV Files\WWE-Data-2013.csv")
wwe2014_df = pd.read_csv(r"C:\Users\RebekahDSK\Desktop\Matplotlib\Winner Wrestling Project\CSV Files\WWE-Data-2014.csv")
wwe2015_df = pd.read_csv(r"C:\Users\RebekahDSK\Desktop\Matplotlib\Winner Wrestling Project\CSV Files\WWE-Data-2015.csv")
wwe2016_df = pd.read_csv(r"C:\Users\RebekahDSK\Desktop\Matplotlib\Winner Wrestling Project\CSV Files\WWE-Data-2016.csv")
```

```python
# Read data and study the results
wwe2013_df.head(5)
wwe2014_df.head(5)
wwe2015_df.head(5)
wwe2016_df.head(5)
```

```python
# Merge the first two datasets (2013 and 2014) on "Wrestler" so that no data is lost (should be 182 rows)
merge2013_2014 = pd.merge(wwe2013_df, wwe2014_df, on='Wrestler', how = 'outer')

# Check the length of rows with len
len(merge2013_2014)
```

```python
# Show a snapshot of the table with .head()
merge2013_2014.head()
```

```python
# Rename our _x columns to "2013 Wins", "2013 Losses", and "2013 Draws" with .rename, calling out columns =
# Start with dataframe that needs renaming 'merged_wwe_df'
merge2013_2014 = merge2013_2014.rename(columns = {'Wins_x': '2013 Wins',
                                               'Draws_x': '2013 Draws',
                                               'Losses_x': '2013 Losses'})
# Rename our _y columns to "2014 Wins", "2014 Losses", and "2014 Draws"
merge2013_2014 = merge2013_2014.rename(columns = {'Wins_y': '2014 Wins',
                                               'Losses_y': '2014 Losses',
                                               'Draws_y': '2014 Draws'})

```

```python
# Merge our newly combined dataframe with the 2015 dataframe
merge2013_2014_2015 = pd.merge(merge2013_2014, wwe2015_df, on = 'Wrestler', how= 'outer')
merge2013_2014_2015.head(5)
```

```python
# Rename "wins", "losses", and "draws" to "2015 Wins", "2015 Losses", and "2015 Draws"
merge2013_2014_2015 = merge2013_2014_2015.rename(columns = {'Wins': '2015 Wins',
                                                           'Losses': '2015 Losses',
                                                           'Draws': '2015 Draws'})
merge2013_2014_2015.head(5)
```

```python
# Merge our newly combined dataframe with the 2016 dataframe

merge2013_2014_2015_2016 = pd.merge(merge2013_2014_2015, wwe2016_df, on ='Wrestler', how= 'outer')
merge2013_2014_2015_2016.head(5)
```

```python
# Rename "wins", "losses", and "draws" to "2016 Wins", "2016 Losses", and "2016 Draws"
merge2013_2014_2015_2016 = merge2013_2014_2015_2016.rename(columns={'Wins': '2016 Wins',
                                                                   'Losses': '2016 Losses',
                                                                   'Draws': '2016 Draws'})
merge2013_2014_2015_2016.head(5)
```

```python
### Part 2 - Time to Calculate!

# When your tables have been merged together into one data frame, calculate the total number of wins, losses, and draws a wrestler has had over the course of their career. Also create a new column that will hold the total matches a wrestler has been in over the course of their career.

# You will need to convert all NaN values to a number so that you can perform these calculations

# We are only interested in those wrestlers who have been with the WWE from 2013 to 2016. You will need to come up with some way of filtering out rows that do not meet these conditions.
    
# Also set the 'Wrestler' column as your key for easier referencing later on.
```

```python
# Replace all NaN values with 0 
merge2013_2014_2015_2016=merge2013_2014_2015_2016.replace(np.nan,0)
merge2013_2014_2015_2016

# Create a new column called "Total Wins" and add up each wrestler's wins per year to fill in the values

# wins = each of the win columns added up
wins = merge2013_2014_2015_2016['2013 Wins'] + merge2013_2014_2015_2016['2014 Wins']\
    + merge2013_2014_2015_2016['2015 Wins'] + merge2013_2014_2015_2016['2016 Wins']
# Within the combined_wrestlers_df, make a column called 'Total Wins' which holds the information we just \
# added up and named 'wins'
merge2013_2014_2015_2016['Total Wins'] = wins
merge2013_2014_2015_2016.head()

# Create a new column called "Total Losses" and add up each wrestler's losses per year to fill in the values

# losses = each of the losses columns added up
losses = merge2013_2014_2015_2016['2013 Losses'] + merge2013_2014_2015_2016['2014 Losses'] \
    + merge2013_2014_2015_2016['2015 Losses'] + merge2013_2014_2015_2016['2016 Losses']
# Within the combined_wrestlers_df, make a column called 'Total Losses' which holds the information we just\
# added up and named 'losses'
merge2013_2014_2015_2016['Total Losses'] = losses
merge2013_2014_2015_2016.head()

# Create a new column called "Total Draws" and add up each wrestler's draws per year to fill in the values

# draws = each of the draws columns added up
draws = merge2013_2014_2015_2016['2013 Draws'] + merge2013_2014_2015_2016['2014 Draws']\
    + merge2013_2014_2015_2016['2015 Draws'] + merge2013_2014_2015_2016['2016 Draws']
# Within the combined_wrestlers_df, make a column called 'Total Draws' which holds the information we just \
# added up and named 'Draws'
merge2013_2014_2015_2016['Total Draws'] = draws
merge2013_2014_2015_2016.head()

# Create a new column called "Total Matches" and add up the total wins, losses, and draws for each wrestler to fill in the values
merge2013_2014_2015_2016['Total Matches'] = wins + losses + draws
merge2013_2014_2015_2016.head()
```

```python
# Create a new dataframe for those wrestlers who have wrestled at least 100 matches,
# have at least one win in 2013, and
# and have at least one win in 2016
# Reference 'Total Matches' column using .loc, greater or equal to 100
OneHun_Matches = merge2013_2014_2015_2016.loc[(merge2013_2014_2015_2016['Total Matches'] >= 100) &
                                            (merge2013_2014_2015_2016['2013 Wins'] >= 1) &
                                          (merge2013_2014_2015_2016['2016 Wins'] >= 1)]
OneHun_Matches.head()
```

```python
# Set the index of this new dataframe to be the wrestlers names
OneHun_Matches = OneHun_Matches.set_index('Wrestler')
OneHun_Matches.head()
```

```python
### Part 3 - Charting Careers

# Store an individual wrestler's wins over time in a variable
# Store that same wrestler's losses over time in a variable as well
# Create a line chart that will plot this wrestler's wins and losses from 2013 to 2016
```

```python
# Collect the user's input to search through our data frame

Wrastler_Name = input("Please enter a wrastler's name:")
```

```python
# Create a series that looks for a wrestler by name and then traces their wins from 2013 to 2016
wrestler_wins = OneHun_Matches.loc[Wrastler_Name, ['2013 Wins', '2014 Wins', '2015 Wins', '2016 Wins']]

# Create a series that looks for a wrestler by name and then traces their losses from 2013 to 2016
wrestler_losses = OneHun_Matches.loc[Wrastler_Name, ['2013 Losses', '2014 Losses', '2015 Losses', '2016 Losses']]
```

```python
# Create a list of the years that we will use as our x axis
years = [2013, 2014, 2015, 2016]
```

```python
# Plot our line that will be used to track a wrestler's wins over the years
plt.plot(years, wrestler_wins, color = 'green', label= 'Wins')

# Plot our line that will be used to track a wrestler's losses over the years
plt.plot(years, wrestler_losses, color = 'red', label = 'Losses')

# Place a legend on the chart in what matplotlib believes to be the "best" location
plt.legend(loc = "best")

plt.title(Wrastler_Name + "'s Career Wins & Losses")
plt.xlabel("Years")
plt.ylabel("Wins & Losses")

# Print our chart to the screen
plt.show()
```
